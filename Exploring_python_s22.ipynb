```python
tdata = np.arange(0, 10, 0.1)
vel = 1.5  # m/s
distance = vel * tdata + 0.1 * np.random.randn(len(tdata))
```

```python
plt.figure()
#plt.plot(tdata, distance, 'o')
plt.errorbar(tdata, distance, yerr=0.1, xerr=0.01, ls='none')
plt.xlabel('time [s]')
plt.ylabel('distance [m]')
```
