```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
```

```python
def beta2(E, U0, m=const.m_e):
    return 2 * m / const.hbar**2 * (U0 - E)

def gamma2(E, U0):
    return ((1 - E / U0) / (E / U0) + (E / U0) / (1 - E / U0) - 2) / 4

def T(L, E, U0, m=const.m_e):
    beta = np.sqrt(beta2(E, U0, m=m))
    return 1 / (np.cosh(beta * L)**2 + gamma2(E, U0) * np.sinh(beta * L)**2)
```

```python
2*np.sqrt(gamma2(6.0, 11.0))

np.sqrt(beta2(6.0 * const.eV, 11.0 * const.eV))

T(0.4e-9, 6.0 * const.eV, 11.0 * const.eV)
```

```python
plt.figure()
x = np.arange(-5, 5, 1e-3)
plt.plot(x, np.sinh(x), label='sinh')
plt.plot(x, np.cosh(x), label='cosh')
plt.grid()
plt.legend()
```

```python
U0 = 10.0 * const.eV # 10 eV in J
L = 1.0e-9  # m
E = np.arange(0, U0, U0 / 1000)
```

```python
plt.figure()
plt.semilogy(E, T(L, E, U0), label='L=1nm')
plt.semilogy(E, T(L / 2, E, U0), label='L=0.5nm')
plt.xlabel('Energy (J)')
plt.ylabel('T')
plt.legend()
```

```python
plt.figure()
E = U0 / 2
L = np.arange(0, 1e-9, 1e-11)
plt.semilogy(L * 1e9, T(L, E, U0))
plt.xlabel('L (nm)')
plt.ylabel('T')
```
